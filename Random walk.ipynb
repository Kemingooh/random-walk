```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulation of a 2D random walk
N = 10**4 # Simulation horizon
alpha = 1/4
beta = 1/4
movement = ['x+1','x-1','y+1','y-1']
pro = [alpha,1/2-alpha,beta,1/2-beta]
X = np.zeros((N,2))
```

```python
for i in range(3):
    for n in range(1,N):
            nextstate = np.random.choice(movement, size=1,p = pro)
            if nextstate == 'x+1':
                X[n][0] = X[n-1][0]+1
                X[n][1] = X[n-1][1]
            elif nextstate == 'x-1': 
                X[n][0] = X[n-1][0]-1
                X[n][1] = X[n-1][1]
            elif nextstate == 'y+1': 
                X[n][0] = X[n-1][0]
                X[n][1] = X[n-1][1]+1
            else : 
                X[n][0]= X[n-1][0]
                X[n][1] = X[n-1][1]-1

    plt.plot(X[:,0],X[:,1])
    plt.title('location of Random walk' + str(i+1))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
alpha = 1/8
beta = 1/8
gama = 1/8
movement = ['x+1','x-1','y+1','y-1','z+1','z-1']
pro = [alpha,1/3-alpha,beta,1/3-beta,gama,1/3-gama]
X = np.zeros((N,3))
```

```python
for i in range(3):
	for n in range(1,N):
			nextstate = np.random.choice(movement, size=1,p = pro)
			if nextstate == 'x+1':
				X[n][0] = X[n-1][0]+1
				X[n][1] = X[n-1][1]
				X[n][2] = X[n-1][2]
			elif nextstate == 'x-1':
				X[n][0] = X[n-1][0]-1
				X[n][1] = X[n-1][1]
				X[n][2] = X[n-1][2]
			elif nextstate == 'y+1':
				X[n][0] = X[n-1][0]
				X[n][1] = X[n-1][1]+1
				X[n][2] = X[n-1][2]
			elif nextstate == 'y-1':
				X[n][0]= X[n-1][0]
				X[n][1] = X[n-1][1]-1
				X[n][2] = X[n-1][2]
			elif nextstate == 'z-1':
				X[n][0] = X[n - 1][0]
				X[n][1] = X[n - 1][1]
				X[n][2] = X[n - 1][2]-1
			else:
				X[n][0] = X[n - 1][0]
				X[n][1] = X[n - 1][1]
				X[n][2] = X[n - 1][2]+ 1
	ax = plt.axes(projection='3d')
	plt.plot(X[:,0],X[:,1],X[:,2])
	plt.title('location of Random walk' + str(i+1))
	plt.xlabel('x')
	plt.ylabel('y')
	plt.show()
```
